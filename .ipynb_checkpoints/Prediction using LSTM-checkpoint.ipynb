```python
import pandas as pd
import numpy as np
import seaborn as sns
import os
import datetime
%pylab inline
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
df=pd.read_csv(".\hourly-energy-consumption\AEP_hourly.csv")
```

```python
df.info()
```

```python
df.describe()
```

```python
df["Date"] = pd.to_datetime(df["Datetime"]).dt.date
df["Time"] = pd.to_datetime(df["Datetime"]).dt.time
```

```python
df.head()
```

```python
figure(figsize=(10, 6))
sns.distplot(df['AEP_MW'])
```

```python
df['Year']=pd.to_datetime(df['Date']).dt.year
```

```python
df.head()
```

```python
df['Year'].unique()
```

#### Relation between Year and Energy

```python
figure(figsize=(10, 6))
sns.lineplot(x=df['Year'],y=df['AEP_MW'],data=df)
```

#### To find the best ft line we will plot the regression

```python
figure(figsize=(15, 8))
sns.jointplot(x=df["Year"],y=df["AEP_MW"],data=df,kind="reg")
```

#### We will find the kernel density estimate so we can see that were the data has most concentration 

```python
figure(figsize=(15, 8))
sns.jointplot(x=df["Year"],y=df["AEP_MW"],data=df,kind="kde")
```

#### Relation between Time and Energy

```python
figure(figsize=(10, 6))
sns.lineplot(x=df['Time'],y=df['AEP_MW'],data=df)
```

```python
dataset = df
dataset["Month"] = pd.to_datetime(df["Datetime"]).dt.month
dataset["Year"] = pd.to_datetime(df["Datetime"]).dt.year
dataset["Date"] = pd.to_datetime(df["Datetime"]).dt.date
dataset["Time"] = pd.to_datetime(df["Datetime"]).dt.time
dataset["Week"] = pd.to_datetime(df["Datetime"]).dt.week
dataset["Day"] = pd.to_datetime(df["Datetime"]).dt.day_name()
dataset = df.set_index("Datetime")
dataset.index = pd.to_datetime(dataset.index)
dataset.head()
```

```python
fig = plt.figure()

ax1= fig.add_subplot(311)
ax2= fig.add_subplot(312)
ax3= fig.add_subplot(313)
ax1.

style.use('ggplot')

y_2004 = dataset["2004"]["AEP_MW"].to_list()
x_2004 = dataset["2004"]["Date"].to_list()
ax1.plot(x_2004,y_2004, color="green", linewidth=1.7,xlable="ssss")


y_2005 = dataset["2005"]["AEP_MW"].to_list()
x_2005 = dataset["2005"]["Date"].to_list()
ax2.plot(x_2005, y_2005, color="green", linewidth=1)


y_2006 = dataset["2006"]["AEP_MW"].to_list()
x_2006 = dataset["2006"]["Date"].to_list()
ax3.plot(x_2006, y_2006, color="green", linewidth=1)


plt.rcParams["figure.figsize"] = (30,20)
plt.suptitle("Energy consumption ",fontsize="40")
plt.xlabel("Date")
plt.ylabel("Energy in MW")
plt.grid(True, alpha=1)
plt.legend()
```
